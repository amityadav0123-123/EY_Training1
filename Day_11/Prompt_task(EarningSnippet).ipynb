```python
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key="YOUR_GROQ_API_KEY"
)

prompt = PromptTemplate(
    template="""
You are a financial analyst.

Summarize the following earnings call snippet in 3 bullet points.

Snippet:
{text}

Summary:
""",
    input_variables=["text"]
)

chain = prompt | llm

response = chain.invoke({
    "text": """
Revenue increased by 18% year-over-year due to strong cloud demand.
Operating margin improved from 12% to 15%.
Management expects continued growth next quarter.
"""
})

print(response.content)
```

```python
from rouge_score import rouge_scorer

reference_summary = """
Revenue increased 18%.
Operating margin improved.
Management expects continued growth.
"""

generated_summary = """
Revenue grew by 18%.
Margins improved.
Company expects future growth.
"""

scorer = rouge_scorer.RougeScorer(
    ['rougeL'],
    use_stemmer=True
)

scores = scorer.score(
    reference_summary,
    generated_summary
)

print(scores['rougeL'])
```

```python
tickets = [
    "I was charged twice for my subscription.",
    "My application crashes when I upload a file.",
    "Please return my payment.",
    "What are your support timings?",
    "This issue has been unresolved for two weeks."
]
```

```python
classifier_prompt = """
You are a customer support expert.

Classify the ticket into one of these categories:

1. Billing
2. Technical
3. Refund
4. General
5. Escalate

Think step-by-step:

Step 1: Understand customer issue.
Step 2: Identify intent.
Step 3: Match category.

Return output in this format:

Reasoning:
<reason>

Category:
<label>

Ticket:
{ticket}
"""
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    template=classifier_prompt,
    input_variables=["ticket"]
)

chain = prompt | llm
```

```python
response = chain.invoke(
    {
        "ticket":
        "I was charged twice for my monthly plan."
    }
)

print(response.content)
```
