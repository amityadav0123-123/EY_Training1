````python
# Step-1: Import required packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# step-2: Load dataset
df = sns.load_dataset("geyser")
df.head()
````

````python
# shape means number of row and columns.here total 272 row and 3 column
df.shape
````

````python
# this shows datatpes of each column
# duration and waiting is numerical column and kind is categorical column
df.dtypes
````

````python
df.describe()
# This will show statistics of numerical columns like mean, std.
````

````python
df.isnull().sum()
# there is no null values so imputation is not required.
````

````python
# numpy operation
mean_duration=np.mean(df['duration'])
mean_duration
````

````python
# apply group by
df.groupby('kind')['duration'].mean()
````

````python
df['kind'].value_counts()
# value_counts tell about unique value and its frequency
````

##Visualization Section

````python
# Bar chart for Kind column
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='kind')

plt.title("Count of Eruption Types")
plt.xlabel("Kind")
plt.ylabel("Count")

plt.show()
````

````python
# Histogram for duration columns
plt.figure(figsize=(8,5))

sns.histplot(data=df,x='duration')

plt.title("Distribution of Eruption Duration")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
````

````python
# Scatter Plot
plt.figure(figsize=(8,5))

sns.scatterplot(data=df,x='duration',y='waiting')

plt.title("Duration vs Waiting Time")
plt.xlabel("Duration (minutes)")
plt.ylabel("Waiting Time (minutes)")

plt.show()
````

````python
# Correlation matrix
corr_matrix = df[['duration', 'waiting']].corr()

# Heatmap
plt.figure(figsize=(6,4))

sns.heatmap(
    corr_matrix,
    annot=True,      # Show correlation values
    cmap='coolwarm', # Color theme
    fmt='.2f'        # Display 2 decimal places
)

plt.title("Correlation Heatmap")
plt.show()
````

# Key Insights from the Geyser Dataset


1. Strong Positive Correlation: There is a strong positive relationship between duration and waiting. Longer eruptions are generally followed by longer waiting times.

2. Two Distinct Eruption Types: The histogram and scatter plot reveal two clear groups of eruptions—short and long—which is why the dataset includes the kind column.

3. Long Eruptions Require More Recovery Time: The average waiting time after a long eruption is significantly higher than after a short eruption, suggesting the geyser needs more time before the next eruption.





# Narrative
The Geyser dataset contains information about water eruptions and the time interval between consecutive eruptions. It includes the eruption duration, the waiting time until the next eruption, and the type of eruption (short or long). Analysis of the data reveals a strong positive relationship between eruption duration and waiting time. Longer eruptions are generally followed by longer waiting periods before the next eruption occurs. The dataset also shows two distinct eruption patterns, which can be observed through histograms and scatter plots. Correlation analysis indicates that the numerical variables are highly related. This dataset is useful for understanding data distributions, relationships between variables, and clustering behavior in real-world observations.
